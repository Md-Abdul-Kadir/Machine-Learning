```python
#importing library
import numpy as np;
import matplotlib.pyplot as plt
import pandas as pd

#import data set
dataSet=pd.read_csv("homeprice.csv")

#print dataset
dataSet
dataSet.head(4)             #fist 4 row
dataSet.shape               #shape of dataSet
dataSet.isnull().sum()      #finding null valus from dataSet


#Visualisation Data

plt.scatter(dataSet.Area , dataSet.Price)
plt.xlabel('Areaqure feet')
plt.ylabel("Price in Taka")
plt.title("Home price in dhaka")



from sklearn.model_selection import train_test_split

x=dataSet[['Area']]
y=dataSet['Price']

xTrain,xTest, yTrain, yTest = train_test_split(x,y,test_size=0.3, random_state=1)

#using linear Regression
from sklearn.linear_model import LinearRegression
RegressionObject=LinearRegression()

#Model Trainig
RegressionObject.fit(xTrain, yTrain)


#value Prediction 
RegressionObject.predict(xTest)      #testing from my data
RegressionObject.predict([[2000]])   #testing random data

```

```python
#how linearRegressin works
#drewing best fit line 

#dataset scatter data
plt.scatter(dataSet.Area , dataSet.Price)
plt.xlabel('Areaqure feet')
plt.ylabel("Price in Taka")
plt.title("Home price in dhaka")

#Best fit line
plt.plot(dataSet.Area , RegressionObject.predict(dataSet[["Area"]]))
```
